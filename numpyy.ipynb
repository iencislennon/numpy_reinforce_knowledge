```python
ages = np.array([[21, 17, 19, 20, 16, 30, 18, 65],
                 [39, 22, 15, 99, 18, 19, 20, 21]])
teens = ages[ages < 18]
print(teens)
uncs = ages[((ages > 30) & (ages < 50)) | (ages<4)]

print(uncs)
```

```python
adults = np.where(ages >= 18, ages, np.nan)
adults
```
